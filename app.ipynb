```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./data.csv")
```

```python
df["Birds (B_CF_Int crop)"] = df["Birds (B_CF_Int crop)"].astype(str)
df["Country"] = df["Country"].astype(str)


plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(df["Country"], df["Birds (B_CF_Int crop)"])
plt.xlabel("Country")
plt.ylabel("Birds (B_CF_Int crop)")
plt.title("Birds (B_CF_Int crop) by Country")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```
